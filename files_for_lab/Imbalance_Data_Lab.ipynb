```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Loading the dataset and checking the variables
data = pd.read_csv('customer_churn.csv')
data.head(30)
```

```python
# Checking data shape
data.shape
```

```python
# Check for all the columns in the dataset
list(data.columns.values)
```

```python
# Checking for each columns unique values
for column in data.columns:
    unique_values = data[column].unique()
    print(f'{column}: {unique_values}')
```

```python
# Confirming the spaces for the one caracter digit
data['tenure'].unique()
```

```python
# Rename columns
column_name_mapping = {'gender': 'Gender','tenure': 'Tenure','customerID': 'CustomerID'}
data.rename(columns=column_name_mapping, inplace=True)
data.head()
```

```python
# Check the number of duplicates
duplicate_count = data.duplicated().sum()
print(f"Number of duplicate records: {duplicate_count}")
```

```python
# Checking for data types
data.dtypes
```

```python
# Identifying how many null values each column has
null_counts = data.isnull().sum()
null_counts
```

```python
# 2 and 3. Splitting tha dataset in X/y 
X = data[['Tenure', 'SeniorCitizen', 'MonthlyCharges']]
y = data['Churn']
```

```python
# Checking my data
X.head()
```

```python
# Checking for the target
y.head()
```

```python
# Checking for correlation between variables
correlations_matrix = X.corr()
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

```python
# look at the scale and distribution of values
sns.displot(X['Tenure'], kde=True)
plt.show()

sns.displot(X['SeniorCitizen'], kde=True)
plt.show()

sns.displot(X['MonthlyCharges'], kde=True)
plt.show()
```

```python
# Train-test-split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
# All features are numeric, there is no need to split into _num and _cat and I can straight forward scale my X-test also
#Scalling my X_train
transformer = StandardScaler().fit(X_train)
X_std_train = transformer.transform(X_train)
X_std_train = pd.DataFrame(X_std_train,columns=X.columns)

# Scalling, but not re-fitting, my X_test
X_std_test = transformer.transform(X_test)
X_std_test = pd.DataFrame(X_std_test,columns=X.columns)
```

```python
X_std_train.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
# The scalling has reset the index for X_train_scaled and for that reason we also have to rese
# so we need this for y_test as well...
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
y_train.head()
y_test.head()
```

```python
# Checking for imbalance of Target values
y.value_counts()
```

```python
#Checking for imbalance
balance_target = pd.value_counts(data['Churn'])
sns.barplot(x=balance_target.index, y=balance_target.values, palette="Set2")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.title("Class Distribution of Churn")
plt.show()
```

```python
# 4. Build the logistic regression model.
from sklearn.linear_model import LogisticRegression

classification = LogisticRegression(random_state =0, solver='lbfgs').fit(X_std_train,y_train)
classification.score(X_std_test, y_test)
```

```python
# 5. Evaluate the Model
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

predictions = classification.predict(X_std_test)

precision = precision_score(y_test, predictions,pos_label='Yes')
recall = recall_score(y_test, predictions,pos_label='Yes')
f1 = f1_score(y_test, predictions,pos_label='Yes')

# Print the results
print("Precision: ", precision)
print("Recall: ", recall)
print("F1-score: ", f1)
```

```python
# Check the values in test
y_test.value_counts()
```

```python
# Check the predicted values
pd.Series(predictions).value_counts()
```

```python
# Condusion _matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
# 6. Even a simple model will give us more than 70% accuracy. Why? That is a result from using highly imbalanced data. The model 
# will attribute by default a guess for the majority of the two values that is 'no' and eventually overlap the results. (check value counts test and predictions)
# Observing the confuion matrix, Precision, Recall and F1-score we can conclude that 0.78 accuracy of the model is quite 
# relative when observing deeper the results
```

```python
# 7. Synthetic Minority Oversampling TEchnique (SMOTE) is an over sampling technique based on nearest neighbors that adds
# new points between existing points. Apply imblearn.over_sampling.SMOTE to the dataset. 
# Build and evaluate the logistic regression model. Is it there any improvement?
```

```python
# Logistic regression trhough SMOTE
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=100, k_neighbors=3) 
X_train_SMOTE,y_train_SMOTE = sm.fit_resample(X_std_train,y_train)
```

```python
X_train_SMOTE.shape
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_SMOTE, y_train_SMOTE)
predictions1 = LR.predict(X_std_test)

print("precision: ",precision_score(y_test,predictions1,pos_label='Yes'))
print("recall: ",recall_score(y_test,predictions1,pos_label='Yes'))
print("f1: ",f1_score(y_test,predictions1,pos_label='Yes'))
```

```python
# Checking the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions1)
```
